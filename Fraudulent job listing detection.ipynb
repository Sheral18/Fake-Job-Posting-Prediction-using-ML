```python
Y = main_df.iloc[:,-1]
X = main_df.iloc[:,:-1]

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```
